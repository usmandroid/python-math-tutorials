# A primer on Numpy Multiplication

```python
import numpy as np
```

```python
print('================')
print('a*b')
print('')
a = np.ceil(10*np.random.randn(3,))
b= np.ceil(10*np.random.randn(3,))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(1,3))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
print(a*b)
print('')
print('================')
print('================')
print('a.dot(b)')
print('')
a = np.ceil(10*np.random.randn(3,))
b= np.ceil(10*np.random.randn(3,))
print(a)
print(b)
print(a.dot(b))
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
# print(a.dot(b))
print("# ValueError: shapes (3,1) and (3,1) not aligned: 1 (dim 1) != 3 (dim 0)")
print('')
a = np.ceil(10*np.random.randn(1,3))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
print(a.dot(b))
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
print(a.dot(b))
print('')
print('================')
print('================')
print('a@b')
print('')
a = np.ceil(10*np.random.randn(3,))
b= np.ceil(10*np.random.randn(3,))
print(a)
print(b)
print(a@b)
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
# print(a@b)
print("matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 3 is different from 1)")
print('')
a = np.ceil(10*np.random.randn(1,3))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
print(a@b)
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
print(a@b) 
print('')
print('================')
```

# Matrix-Matrix and Matrix-Array Multiplication

```python
print('================')
print('a*b')
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,2))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(2,3))
print(a)
print(b)
# print(a*b)
print("ValueError: operands could not be broadcast together with shapes (3,2) (2,3)")
print('')
a = np.ceil(10*np.random.randn(2,3))
b= np.ceil(10*np.random.randn(2,3))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
print(a*b)
print('')
a = np.ceil(10*np.random.randn(2,3))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
print(a*b)
print('')
print('================')
print('================')
print('a.dot(b)')
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,2))
print(a)
print(b)
# print(a.dot(b))
print("ValueError: shapes (3,2) and (3,2) not aligned: 2 (dim 1) != 3 (dim 0)")
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(2,3))
print(a)
print(b)
print(a.dot(b))
print('')
a = np.ceil(10*np.random.randn(2,3))
b= np.ceil(10*np.random.randn(2,3))
print(a)
print(b)
# print(a.dot(b))
print("ValueError: shapes (2,3) and (2,3) not aligned: 3 (dim 1) != 2 (dim 0)")
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
# print(a.dot(b))
print("ValueError: shapes (3,2) and (3,1) not aligned: 2 (dim 1) != 3 (dim 0)")
print('')
a = np.ceil(10*np.random.randn(3,1))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
print(a.dot(b))
print('')
print('================')
print('================')
print('a@b')
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,2))
print(a)
print(b)
# print(a@b)
print("ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 3 is different from 2)")
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(2,3))
print(a)
print(b)
print(a@b)
print('')
a = np.ceil(10*np.random.randn(2,3))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
# print(a@b)
print("ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 1 is different from 3)")
print('')
a = np.ceil(10*np.random.randn(3,2))
b= np.ceil(10*np.random.randn(3,1))
print(a)
print(b)
# print(a@b)
print("ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 3 is different from 2)")
print('')
a = np.ceil(10*np.random.randn(2,3))
b= np.ceil(10*np.random.randn(1,3))
print(a)
print(b)
# print(a@b) 
print("ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 1 is different from 3)")
print('')
print('================')
```
